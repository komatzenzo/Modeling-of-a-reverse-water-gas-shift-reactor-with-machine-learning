```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from io import StringIO
from sklearn.model_selection   import train_test_split
from sklearn.preprocessing     import StandardScaler
from sklearn.ensemble          import RandomForestRegressor
from sklearn.multioutput       import MultiOutputRegressor
from sklearn.pipeline          import Pipeline
from sklearn.metrics           import mean_squared_error, r2_score

df = pd.read_csv('data.csv', delimiter=';')
nan_locations = df.isna().sum()
```

```python
#Feature / Target Split and Train/Test Split

X = df[['Tout','pset','H2toCO2','GHSV']].values
y = df[['XCO2','SCO']].values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=None
)
print("Train samples:", X_train.shape[0], "| Test samples:", X_test.shape[0])
# Standardize
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)

feature_names = ['Tout', 'pset', 'H2toCO2', 'GHSV']
print("\nScaled feature statistics train set:")
print(pd.DataFrame(X_train, columns=feature_names).describe())
```

```python
# pip install scikit-optimize scipy

import numpy as np
import matplotlib.pyplot as plt

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, r2_score

from skopt import BayesSearchCV
from skopt.space import Categorical, Real
import optuna
from sklearn.model_selection import cross_val_score
from sklearn.base import clone
import numpy as np
from matplotlib import rcParams
from scipy.ndimage import gaussian_filter1d

pipe = Pipeline([
    ('mlp', MLPRegressor(
        solver='adam',
        max_iter=5000,
        early_stopping=True,
        n_iter_no_change=50,
        validation_fraction=0.2,
        #random_state=42 #Immer gleicher Seed
    ))
])

def objective(trial):
    first = trial.suggest_int('first_layer_neurons', 10, 100, step=10)
    second = trial.suggest_int('second_layer_neurons', 0, 100, step=10)

    if second == 0:
        hidden = (first,)
    else:
        hidden = (first, second)

    params = {
        'mlp__hidden_layer_sizes': hidden,
        'mlp__activation': trial.suggest_categorical('mlp__activation', ['relu', 'tanh']),
        'mlp__alpha': trial.suggest_float('mlp__alpha', 1e-6, 1e-1, log=True),
        'mlp__learning_rate_init': trial.suggest_float('mlp__learning_rate_init', 1e-4, 1e-1, log=True),
        'mlp__beta_1': trial.suggest_float('mlp__beta_1', 0.85, 0.99),
        'mlp__beta_2': trial.suggest_float('mlp__beta_2', 0.90, 0.999),
    }
    model = clone(pipe)
    model.set_params(**params)
    scores = cross_val_score(
        model,
        X_train,
        y_train,
        cv=5,
        scoring='neg_mean_squared_error',
        n_jobs=-1
    )
    return scores.mean()

study = optuna.create_study(direction='maximize', sampler=optuna.samplers.TPESampler(seed=42))
study.optimize(objective, n_trials=10, show_progress_bar=True)
raw_best_params = study.best_params
best_score = study.best_value
first = raw_best_params['first_layer_neurons']
second = raw_best_params['second_layer_neurons']
hidden = (first,) if second == 0 else (first, second)
best_params = {k: v for k, v in raw_best_params.items()
               if k not in ('first_layer_neurons', 'second_layer_neurons')}
best_params['mlp__hidden_layer_sizes'] = hidden

best_estimator = clone(pipe)
best_estimator.set_params(**best_params)
best_estimator.fit(X_train, y_train)

class OptunaResult:
    def __init__(self, best_estimator, best_params, best_score):
        self.best_estimator_ = best_estimator
        self.best_params_ = best_params
        self.best_score_ = best_score  # neg MSE
    def fit(self, X, y):
        self.best_estimator_.fit(X, y)
        return self
    def predict(self, X):
        return self.best_estimator_.predict(X)

opt = OptunaResult(best_estimator, best_params, best_score)
if y_train.ndim > 1 and y_train.shape[1] == 1:
    y_train = y_train.ravel()

opt.fit(X_train, y_train)
print("Best params:", opt.best_params_)
print("Best CV score (neg MSE):", opt.best_score_)

y_train_pred = np.clip(opt.predict(X_train), 0.0, 100.0)
y_test_pred  = np.clip(opt.predict(X_test), 0.0, 100.0)
if y_train.ndim > 1 and y_train_pred.ndim == 1:
    y_train_pred = y_train_pred.reshape(-1, 1)
if y_test.ndim > 1 and y_test_pred.ndim == 1:
    y_test_pred = y_test_pred.reshape(-1, 1)

y_train = np.asarray(y_train)
y_test  = np.asarray(y_test)
y_train_pred = np.asarray(y_train_pred)
y_test_pred  = np.asarray(y_test_pred)

if y_train.shape != y_train_pred.shape:
    raise ValueError(f"Shape mismatch between y_train {y_train.shape} and y_train_pred {y_train_pred.shape}")
if y_test.shape != y_test_pred.shape:
    raise ValueError(f"Shape mismatch between y_test {y_test.shape} and y_test_pred {y_test_pred.shape}")


mse_train_all = np.mean((y_train - y_train_pred) ** 2)
mse_test_all  = np.mean((y_test  - y_test_pred)  ** 2)
rmse_train = np.sqrt(mse_train_all)
rmse_test  = np.sqrt(mse_test_all)
print(f"Train RMSE (beide Ziele): {rmse_train:.2f}")
print(f"Test  RMSE (beide Ziele): {rmse_test:.2f}")

r2_train = r2_score(y_train, y_train_pred, multioutput='variance_weighted')
r2_test  = r2_score(y_test,  y_test_pred,  multioutput='variance_weighted')
print(f"Train R²: {r2_train:.2f}")
print(f"Test  R²: {r2_test:.2f}")

# Print overall MSE 
mse_train = mse_train_all
mse_test  = mse_test_all
print(f"Train MSE: {mse_train:.2f}")
print(f"Test  MSE: {mse_test:.2f}")
# ========= MAE =========
mae_train_all = np.mean(np.abs(y_train - y_train_pred))
mae_test_all  = np.mean(np.abs(y_test  - y_test_pred))
print(f"Train MAE (beide Ziele): {mae_train_all:.2f}")
print(f"Test  MAE (beide Ziele): {mae_test_all:.2f}")

for i, name in enumerate(['X(CO2)', 'S(CO)']):
    mse_i = mean_squared_error(y_test[:, i], y_test_pred[:, i])
    rmse_i = np.sqrt(mse_i)
    r2_i   = r2_score(y_test[:, i], y_test_pred[:, i])
    print(f"Test RMSE {name}: {rmse_i:.2f} | R² {name}: {r2_i:.2f}")

# ========= Loss =========
rcParams['text.usetex'] = False
rcParams['font.family'] = 'serif'

best_mlp = opt.best_estimator_.named_steps['mlp']
loss = np.asarray(best_mlp.loss_curve_, dtype=float)
epochs = np.arange(1, loss.size + 1)
if loss.size > 5:
    loss_smooth = gaussian_filter1d(loss, sigma=2)
else:
    loss_smooth = loss

plt.figure(figsize=(6,4))
plt.plot(epochs, loss, label='Training $\\mathcal{L}_{\\mathrm{MSE}}$', color='C0', lw=1.5, alpha=0.5)
plt.scatter([epochs[-1]], [loss[-1]], color='C0', zorder=5)

plt.xlabel('Epoch')
plt.ylabel('Loss $\\mathcal{L}_{\\mathrm{MSE}}$')
plt.title('Artificial neural network training loss')
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

# ========= Parity Plots =========
titles = [r'$X_{\mathrm{CO_2}}$', r'$S_{\mathrm{CO}}$']
plt.figure(figsize=(12,6))
for i in range(2):
    plt.subplot(1,2,i+1)
    plt.scatter(y_train[:, i], y_train_pred[:, i], label='Train', alpha=1)
    plt.scatter(y_test[:,  i], y_test_pred[:,  i], label='Test',  alpha=1, marker='^')
    plt.plot([0,100],[0,100], 'k--', lw=1)
    plt.xlim(0,100); plt.ylim(0,100)
    plt.xlabel(f"True {titles[i]} [%]"); plt.ylabel(f"Predicted {titles[i]} [%]")
    plt.title(f"{titles[i]}: Parity")
    plt.legend()
plt.tight_layout(); plt.show()
```

```python
# SHAP Analysis
import shap
import warnings
warnings.filterwarnings('ignore')
if 'best_estimator' in globals():
    model = best_estimator
elif 'opt' in globals() and hasattr(opt, 'best_estimator_'):
    model = opt.best_estimator_
else:
    raise NameError("best_estimator not found")
X_all = np.vstack([X_train, X_test])  # scaled
df_all = pd.DataFrame(scaler.inverse_transform(X_all), columns=feature_names)
X_for_explainer = X_all
n_background = min(100, X_for_explainer.shape[0])
rng = np.random.default_rng(42)
background_idx = rng.choice(X_for_explainer.shape[0], n_background, replace=False)
background = X_for_explainer[background_idx]

def make_predict_fn(out_i):
    def predict_fn(x):
        x = np.asarray(x)
        if x.ndim == 1:
            x = x.reshape(1, -1)
        preds = model.predict(x)
        preds = np.asarray(preds)
        if preds.ndim == 1:
            preds = preds.reshape(-1, 1)
        return preds[:, out_i]
    return predict_fn

target_names = [r'$X_{\mathrm{CO_2}}$', r'$S_{\mathrm{CO}}$']
shap_values_per_target = []
explainers_per_target = []
base_scalar_per_target = []

for out_i in range(len(target_names)):
    predict_fn = make_predict_fn(out_i)
    shap_vals = None
    explainer = None
    for ns in (100, 50, 25):
        try:
            explainer = shap.KernelExplainer(predict_fn, background)
            raw_vals = explainer.shap_values(X_for_explainer, nsamples=ns)
            shap_vals = raw_vals
            break
        except Exception as e:
            print(f"KernelExplainer nsamples={ns} failed for target {out_i}: {e}")
            shap_vals = None
            explainer = None
            continue
    if shap_vals is None:
        raise RuntimeError(f"SHAP computation failed for target {out_i} with all tried nsamples")

    # Normalise
    shap_arr = np.asarray(shap_vals)
    if shap_arr.ndim == 1:
        shap_arr = shap_arr.reshape(-1, 1)
    elif shap_arr.ndim == 2:
        n_samp = X_for_explainer.shape[0]
        n_feat = X_for_explainer.shape[1]
        if shap_arr.shape[0] == n_feat and shap_arr.shape[1] == n_samp:
            shap_arr = shap_arr.T
    elif shap_arr.ndim == 3:
        shap_arr = shap_arr.reshape(shap_arr.shape[1], shap_arr.shape[2])

    if explainer is not None and hasattr(explainer, 'expected_value'):
        exp_val = explainer.expected_value
        try:
            base_scalar = float(exp_val)
        except Exception:
            base_scalar = float(np.asarray(exp_val).mean())
    else:
        base_scalar = float(predict_fn(background).mean())

    shap_values_per_target.append(shap_arr)
    explainers_per_target.append(explainer)
    base_scalar_per_target.append(base_scalar)

for i, name in enumerate(target_names):
    try:
        shap_vals = shap_values_per_target[i]
        n_samples_shap = shap_vals.shape[0]
        data_for_expl = df_all.iloc[:n_samples_shap].reset_index(drop=True).values
        base_val = base_scalar_per_target[i]
        expl = shap.Explanation(values=shap_vals, base_values=base_val, data=data_for_expl, feature_names=feature_names)
        shap.plots.bar(expl, max_display=len(feature_names), show=False)
        plt.title(f'Global Feature Importance — {name}')
        plt.tight_layout(); plt.show()
    except Exception as e:
        print(f"Global importance plot failed for {name}: {e}")

for idx, name in enumerate(target_names):
    try:
        shap_vals = shap_values_per_target[idx]
        n_samples_shap = shap_vals.shape[0]
        df_all_plot = df_all.iloc[:n_samples_shap].reset_index(drop=True)
        shap.summary_plot(shap_vals, df_all_plot, feature_names=feature_names, show=False)
        plt.title(f"{name} — SHAP Summary Plot")
        plt.show()
    except Exception as e:
        print(f"SHAP summary plot failed for {name}: {e}")
        continue
    try:
        mean_abs = np.abs(shap_vals).mean(axis=0)
        top_idx = int(mean_abs.argmax())
        top_feat = feature_names[top_idx]

        feat_to_plot = 'Tout' if 'Tout' in feature_names else top_feat
        feat_idx = feature_names.index(feat_to_plot)
        if 'pset' in df_all_plot.columns:
            color_array = df_all_plot['pset'].values
            used_color = 'pset'
        elif 'p_set' in df_all_plot.columns:
            color_array = df_all_plot['p_set'].values
            used_color = 'p_set'
        elif 'Tout' in df_all_plot.columns:
            color_array = df_all_plot['Tout'].values
            used_color = 'Tout'
        else:
            color_array = shap_vals[:, top_idx]
            used_color = f"shap({top_feat})"
        vals = np.asarray(shap_vals)
        n_samp = vals.shape[0]
        base_vals = np.full((n_samp,), float(base_scalar_per_target[idx]))
        feat_vals = vals[:, feat_idx].reshape(n_samp, 1)
        feat_data = df_all_plot[[feat_to_plot]].values.reshape(n_samp, 1)
        color_array = np.asarray(color_array).reshape(-1)[:n_samp]
        feat_expl = shap.Explanation(values=feat_vals, base_values=base_vals, data=feat_data, feature_names=[feat_to_plot])
        shap.plots.scatter(feat_expl, color=color_array, show=False)
        plt.title(f"{name}: {feat_to_plot} colored by {used_color}")
        plt.tight_layout(); plt.show()



        if feat_to_plot == 'Tout':
            y_all = np.vstack([y_train, y_test])[:n_samp]
            try:
                corr_feat = np.corrcoef(df_all_plot[feat_to_plot].values, vals[:, feature_names.index(feat_to_plot)])[0, 1]
                corr_xco2 = np.corrcoef(y_all[:, 0], vals[:, feature_names.index(feat_to_plot)])[0, 1]
                corr_sco  = np.corrcoef(y_all[:, 1], vals[:, feature_names.index(feat_to_plot)])[0, 1]
                print(f"Correlation SHAP-{feat_to_plot}: {corr_feat:.2f} | vs X_CO2: {corr_xco2:.2f} | vs S_CO: {corr_sco:.2f}")
            except Exception as e:
                print('Correlation computation failed:', e)
    except Exception as e:
        print(f"Per-feature SHAP scatter failed for {name}: {e}")
```

## Comparison of ANN Model, Thermodynamic data and Experimental data

```python
target_ghsv = 20000
target_pset = 1
target_h2co2 = 3.0
tout_min, tout_max = 550.0, 950.0

mask = (
    (df['Tout'] >= tout_min) & (df['Tout'] <= tout_max)
    & np.isclose(df['H2toCO2'], target_h2co2)
    & np.isclose(df['GHSV'], target_ghsv)
    & np.isclose(df['pset'], target_pset)
)
df_sel = df[mask].sort_values('Tout')
if df_sel.empty:
    mask_h2_tout = (df['Tout'] >= tout_min) & (df['Tout'] <= tout_max) & np.isclose(df['H2toCO2'], target_h2co2)
    if mask_h2_tout.any():
        df_h2 = df[mask_h2_tout].copy()
        df_h2['dist'] = (np.abs(df_h2['pset'] - target_pset)
                         + np.abs(df_h2['GHSV'] - target_ghsv))
        df_sel = df_h2.sort_values(['dist', 'Tout']).drop(columns='dist')
    else:
        df_in_tout = df[(df['Tout'] >= tout_min) & (df['Tout'] <= tout_max)].copy()
        if df_in_tout.empty:
            closest_h2 = df['H2toCO2'].iloc[(np.abs(df['H2toCO2'] - target_h2co2)).argmin()]
            df_sel = df.sort_values('Tout')
        else:
            closest_h2 = df_in_tout['H2toCO2'].iloc[(np.abs(df_in_tout['H2toCO2'] - target_h2co2)).argmin()]
            df_sel = df_in_tout.sort_values('Tout')

Tout_exp = df_sel['Tout'].values
y_exp = df_sel[['XCO2', 'SCO']].values
X_sel = df_sel[['Tout', 'pset', 'H2toCO2', 'GHSV']].values
mlp = best_estimator.named_steps['mlp']

X_sel_scaled = scaler.transform(X_sel)
y_ann_sel_raw = mlp.predict(X_sel_scaled)
y_ann_sel = np.clip(y_ann_sel_raw, 0.0, 100.0)

Tout_grid = np.linspace(tout_min, tout_max, 200)
h2co2_val = target_h2co2
pset_val = target_pset
ghsv_val = target_ghsv

X_grid = np.column_stack([
    Tout_grid,
    np.full_like(Tout_grid, pset_val),
    np.full_like(Tout_grid, h2co2_val),
    np.full_like(Tout_grid, ghsv_val)
])
X_grid_scaled = scaler.transform(X_grid)
y_ann_grid_raw = mlp.predict(X_grid_scaled)
y_ann_grid = np.clip(y_ann_grid_raw, 0.0, 100.0)

# === Load equilibrium data (sthermo_data.xlsx) ===
y_eq_pred = None
tout_eq = None
try:
    eq_df = pd.read_excel('thermo_data.xlsx')
    # Ensure expected columns exist
    required_cols = {'RATIOH2CO2','PRESBAR','Tout','XCO2','SCO'}
    if required_cols.issubset(eq_df.columns):
        eq_mask = (
            np.isclose(eq_df['RATIOH2CO2'], h2co2_val)
            & np.isclose(eq_df['PRESBAR'], pset_val)
            & (eq_df['Tout'] >= tout_min) & (eq_df['Tout'] <= tout_max)
        )
        eq_sel = eq_df[eq_mask].sort_values('Tout')
        if not eq_sel.empty:
            tout_eq = eq_sel['Tout'].to_numpy()
            y_eq_pred = np.column_stack([eq_sel['XCO2'].to_numpy(),
                                         eq_sel['SCO'].to_numpy()])
    else:
        print("Equilibrium file missing required columns, skipping equilibrium curve.")
except FileNotFoundError:
    print("sthermo_data.xlsx not found; skipping equilibrium overlay.")

titles_plot = [r'$X_{\mathrm{CO_2}}$ [%]', r'$S_{\mathrm{CO}}$ [%]']

fig, axes = plt.subplots(1, 2, figsize=(12, 4), sharey=True)
for i, ax in enumerate(axes):
    if y_eq_pred is not None and tout_eq is not None:
        ax.plot(tout_eq, y_eq_pred[:, i], linestyle='--', color='dodgerblue', label='Equilibrium')
    ax.plot(Tout_grid, y_ann_grid[:, i], linestyle='-', label='ANN model', color='C1', alpha=0.9)
    if Tout_exp.size:
        ax.scatter(Tout_exp, y_exp[:, i], marker='o', color='C1', zorder=5, label='Experimental')
    ax.set_xlabel('Temperature $T_{out}$')
    if i == 0:
        ax.set_ylabel(titles_plot[i])
    ax.set_title(f"{titles_plot[i]} vs Temperature (GHSV={target_ghsv}, p={target_pset} bara, H2:CO2={h2co2_val})")
    ax.set_xlim(Tout_grid.min(), Tout_grid.max())
    ax.set_ylim(0, 100)
    ax.legend()
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
